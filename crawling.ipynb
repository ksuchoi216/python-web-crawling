```python
import requests
from bs4 import BeautifulSoup
import re


source_url = 'https://m.cafe.daum.net/ourvancouver/Dqvd/'
num = 207

url = source_url+str(num)


req = requests.get(url)
html = req.content
```

```python
soup = BeautifulSoup(html, 'lxml')
# print(soup)

div = soup.find(name='div', attrs={'id': 'article','class': 'tx-content-container'})
rows = div.find_all(name='span')


checker = False
for i in range(0, len(rows)):
    text = rows[i].text
    checking_text = text.strip()
    if checking_text == '1.':
        checker=True

    if checker:
        text = text.replace('\n', "")
        print(text)
    
```
