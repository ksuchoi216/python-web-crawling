```python
import re
import os

# keyword = 'basic'
keyword = 'advanced'
rpath = f'/Users/KC/Documents/mk_{keyword}_500.txt'
wpath = f'/Users/KC/Documents/mk_{keyword}_500_anki.txt'

if keyword == 'basic':
    key = 'B'
elif keyword == 'advanced':
    key = 'A'



with open(wpath, 'w') as fw:
    pass

f = open(wpath, "w")

with open(rpath, 'r') as fr:
    lines = fr.readlines()

    front = []
    back = []
    
    main_num = 0
    texts = []
    isStore = False
    prev_isKor = False
    wline = ''
    for i, line in enumerate(lines):
        line = line.rstrip('\n')

        lis = re.split('\s', line)
        isNum = bool(re.search('[0-9]+.', lis[0]))
        # isEng = bool(re.search('[a-zA-z]+', line))
        now_isKor = bool(re.search('[가-힣]+', line))
        if isNum:  
            sub_num = int(re.findall('[0-9]+', lis[0])[0])
            if sub_num == 1:
                main_num += 1
            
            
            numbering = f'[{key}{main_num}-{sub_num}]'
            lis[0] = numbering
            # print(lis
            store_line = ' '.join(text for text in lis)
            isStore = True
        else:
            store_line = line

        if isStore:
            if prev_isKor is False and now_isKor is False:
                texts.append(store_line)
            elif prev_isKor is False and now_isKor is True:
                texts.append(';')
                texts.append(numbering+' '+store_line)
            elif prev_isKor is True and now_isKor is True:
                texts.append(store_line)
            elif prev_isKor is True and now_isKor is False:
                wline = ' '.join(text for text in texts)
                wline = wline + '\n'
                f.write(wline)
                print(wline)
                wline = ''
                texts = []
                isStore = False
                
        prev_isKor = now_isKor

f.close()
```
